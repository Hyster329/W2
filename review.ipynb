# Cab data

```python
cab = pd.read_csv(r'E:\dataglacier\DataSets\Cab_Data.csv')
cab=cab.sample(frac=1,ignore_index=True)
```

# Since I have randomise the data I assume the data is equally randomisedistributred

```python
# View first twenty thousand
cab=cab.head(20000)
```

```python
import datetime
ex_date= cab['Date of Travel']
year =[]
for i in range(len(ex_date)):
    excel_based_date = datetime.datetime(1899,12,30)
    date = excel_based_date + datetime.timedelta(days=int(ex_date[i]))
    year.append(date.year)
cab['Year of Travel']= year
```

```python
sn.barplot(x='City',y='Cost of Trip', hue='Company',data=cab,ci=None)
plt.xticks(rotation=90)
plt.title('Cost of travel against Cities in  relation to Company')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')

```

```python
ax=sn.barplot(x='City', y='Price Charged',data=cab, ci=None,width=0.9)
plt.xticks(rotation=90)
# for p in ax.patches:
#     ax.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()- 1),rotation=90,
#                 ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), 
#                 textcoords='offset points')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout(pad=0.2)
plt.title('Price Chargered per city ')
```
